```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('words')
from textblob import TextBlob
```

```python
def autocorrect(text):
    blob = TextBlob(text)
    return str(blob.correct())

text = "Ths is a twst."
corrected_text = autocorrect(text)
print(corrected_text)
```
